```python
import cv2

# Load face detection model
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    if not ret:
        print("Camera not working")
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.3,
        minNeighbors=5
    )

    for (x, y, w, h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)

    cv2.imshow("Face Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from fer import FER

# ---------- Overlay Function ----------
def overlay_png(background, overlay, x, y):

    h, w = overlay.shape[:2]

    if y + h > background.shape[0] or x + w > background.shape[1]:
        return

    alpha = overlay[:, :, 3] / 255.0
    overlay_rgb = overlay[:, :, :3]

    for c in range(3):
        background[y:y+h, x:x+w, c] = (
            alpha * overlay_rgb[:, :, c] +
            (1 - alpha) * background[y:y+h, x:x+w, c]
        )

# ---------- Load Persona Images ----------
happy_img = cv2.imread("../persona_filters/happy.png", cv2.IMREAD_UNCHANGED)
sad_img = cv2.imread("../persona_filters/sad.png", cv2.IMREAD_UNCHANGED)
angry_img = cv2.imread("../persona_filters/angry.png", cv2.IMREAD_UNCHANGED)
neutral_img = cv2.imread("../persona_filters/neutral.png", cv2.IMREAD_UNCHANGED)

# ---------- Emotion Detector ----------
emotion_detector = FER()

cap = cv2.VideoCapture(0)

# Bigger camera window
cap.set(3, 1280)
cap.set(4, 720)

while True:

    ret, frame = cap.read()

    if not ret:
        break

    emotions = emotion_detector.detect_emotions(frame)

    if emotions:

        x, y, w, h = emotions[0]["box"]

        emotion, score = max(
            emotions[0]["emotions"].items(),
            key=lambda x: x[1]
        )

        # Draw face box
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)

        # Emotion → Persona mapping
        if emotion == "happy":
            persona = happy_img
        elif emotion == "sad":
            persona = sad_img
        elif emotion == "angry":
            persona = angry_img
        else:
            persona = neutral_img

        persona = cv2.resize(persona,(w,int(h*0.8)))

        y_offset = max(0, y - int(h*0.9))
        x_offset = x

        overlay_png(frame, persona, x_offset, y_offset)

        # Emotion label
        cv2.putText(
            frame,
            emotion,
            (x, y-10),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0,255,0),
            2
        )

        # ---------- 4 Emotion Bars ----------
        emotions_dict = emotions[0]["emotions"]

        bar_x = 20
        bar_y = 40

        selected_emotions = ["angry", "sad", "happy", "neutral"]

        # Emotion colors (BGR)
        emotion_colors = {
            "angry": (0,0,255),     # Red
            "sad": (255,0,0),       # Blue
            "happy": (0,255,255),   # Yellow
            "neutral": (0,255,0)    # Green
        }

        for emotion_name in selected_emotions:

            value = emotions_dict.get(emotion_name, 0)

            percentage = int(value * 100)

            text = f"{emotion_name}: {percentage}%"

            cv2.putText(
                frame,
                text,
                (bar_x, bar_y),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.7,
                (255,255,255),
                2
            )

            # Background bar
            cv2.rectangle(
                frame,
                (180, bar_y-15),
                (420, bar_y),
                (50,50,50),
                -1
            )

            # Colored emotion bar
            cv2.rectangle(
                frame,
                (180, bar_y-15),
                (180 + int(value*240), bar_y),
                emotion_colors[emotion_name],
                -1
            )

            bar_y += 40

    cv2.imshow("AI Emotion Persona Overlay", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
